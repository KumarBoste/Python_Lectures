```python
# SLICING A STRING
## string[start:stop]

string = "Hello World"
string1 = "India is Great"

print(string[:5]) # o/p: Hello
print(string[:-6]) # o/p: Hello
print(string1[9:]) # o/p: Great
print(string1[-5:])

```

```python
# Q. Write a program to print sentence/word is palindrom or not.
'''
Logic:
1. ask user to enter a string.
2. store the string in variable : string
3. reverse the entered string : string1
4. use if else conditions to the string is palindrom or not.

'''
## Step - 1 : Ask user to enter the string
string = input("Enter a String: ")

## Step - 2 : Reverse the entered string : string1
string1 = string[::-1]

## Step - 3 : use if else logics
if string == string1 :
  print("The Given String is Palindrom")
else:
  print("The Given String is not Palindrom")
```

```python
# Q . Segregate user based on Marks Scored
## Condition : If Marks >= 75 : Bright Student else weak Student.

'''
Logic :

1. ask user to enter the marks.
2. store the marks in variable : marks
3. use if else logics
'''

## Step - 1 : Ask user to enter the marks
marks = float(input("Enter the marks: "))

## Step - 2 : Store the marks in variable : marks

## Step - 3  : Use if else Logics
if marks >= 75 :
  print("Bright Student")
else :
  print("Weak Student")
```
