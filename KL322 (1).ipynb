```python
# Q. Write a program to print sentence/word is palindrom or not.
'''
Logic:
1. ask user to enter the string.
2. store the string in variable: string
3. reverse the entered string : string1
4. use if else condition to check the string is palindrom or not.
'''
# Step1: Ask user to enter the string
string = input("Enter the String: ")

# Step2: Reverse the entered string: string1
string1 = string[::-1]

# Step3: Use if else logics
if string == string1:
  print("The Given String is Palindrom")
else:
  print("The Given String is not Palindrom")
```

```python
# Number Guessing Game : With Latency Calculation
import time
import random

start = time.perf_counter()

number = random.randint(1,5)
print(number)

if number == 1:
  print("Number is 1")
elif number == 2:
  print("Number is 2")
elif number == 3:
  print("Number is 3")
elif number == 4:
  print("Number is 4")
elif number == 5:
  print("Number is 5")
else:
  print("No Number")

# Your code that you want to time
end = time.perf_counter() - start
print('{:.6f}s for the calculation'.format(end))
```
