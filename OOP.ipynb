## Concept of Encapsulation

```python
# encapsulation method

class BankAccount:

    # constructor
    def __init__(self, owner, balance=0):
        self.owner = owner            # public attribute/ feature
        self.__balance = balance      # private attribute/ feature

    # Getter method (read-only access to balance)
    def get_balance(self):
        return self.__balance

    # Deposit method with validation
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount  # amount + __balance
            print(f"Deposited: {amount}")
        else:
            print("Invalid deposit amount!")

    # Withdraw method with validation
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount  # amount - __balance
            print(f"Withdrawn: {amount}")
        else:
            print("Insufficient funds or invalid amount!")

# Using the class ---
acc = BankAccount("Chandan", 1000)

acc.deposit(500)
acc.withdraw(300)

print("Current Balance:", acc.get_balance())
```

```python
# Example of Encapsulation

class BankAccount:
    def __init__(self, owner, pin, balance=0):
        self.owner = owner         # public attribute
        self.__pin = pin           # private attribute
        self.__balance = balance   # private attribute

    # Method to verify PIN (private helper method)
    def __verify_pin(self, pin):
        return pin == self.__pin   # comparison

    # Public method – Check balance
    def check_balance(self, pin):
        if self.__verify_pin(pin):
            return f"Available Balance: {self.__balance}"
        else:
            return "Invalid PIN!"

    # Public method – Deposit money
    def deposit(self, pin, amount):
        if not self.__verify_pin(pin):
            return "Invalid PIN!"

        if amount > 0:
            self.__balance += amount
            return f"Deposited {amount}. New balance: {self.__balance}"
        return "Invalid deposit amount."

    # Public method – Withdraw money
    def withdraw(self, pin, amount):
        if not self.__verify_pin(pin):
            return "Invalid PIN!"

        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return f"Withdrawn {amount}. Remaining balance: {self.__balance}"
        return "Insufficient funds or invalid amount."


# Test Case:
account = BankAccount("Chandan", pin=1234, balance=5000)

print(account.check_balance(1234))
print(account.withdraw(1234, 1000))
print(account.deposit(1234, 2000))
print(account.check_balance(1111))
```

```python
# Real time example: Inheritance

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited: {amount}")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print("Insufficient funds!")

    def get_balance(self):
        return self.balance


class SavingsAccount(BankAccount):
    def __init__(self, owner, balance=0, interest_rate=0.05):
        super().__init__(owner, balance)
        self.interest_rate = interest_rate

    def add_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Interest added: {interest}")


class CurrentAccount(BankAccount):
    def __init__(self, owner, balance=0, overdraft_limit=1000):
        super().__init__(owner, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if amount <= self.balance + self.overdraft_limit:
            self.balance -= amount
            print(f"Withdrawn using overdraft: {amount}")
        else:
            print("Exceeded overdraft limit!")


s = SavingsAccount("Chandan", 2000)
s.add_interest()
print("Savings Balance:", s.get_balance())

c = CurrentAccount("Chandan", 1000)
c.withdraw(1500)
print("Current Balance:", c.get_balance())
```
